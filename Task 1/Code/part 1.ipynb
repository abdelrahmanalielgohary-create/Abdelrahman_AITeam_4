```
import numpy as np
```

```
grades_list = [
    [85, 78, 92, 88],
    [70, 76, 80, 65],
    [90, 88, 94, 91],
    [60, 65, 58, 62],
    [100, 95, 98, 97]]
```

```
grades = np.array(grades_list, dtype=float)
```

```
print("Shape:", grades.shape)
```

```
mean_per_student = grades.mean(axis=1)
print("Mean per student:", mean_per_student)
```

```
mean_per_subject = grades.mean(axis=0)
print("Mean per subject:", mean_per_subject)
```

```
high_avg_mask = mean_per_student > 85
students_high_avg = grades[high_avg_mask]
print("Mask of students with avg > 85:", high_avg_mask)
print("Students (rows) with avg > 85:\n", students_high_avg)
```

```
bonus_grades = grades + 5
print("Grades after +5 bonus:\n", bonus_grades)
```

```
col_min = grades.min(axis=0)
col_max = grades.max(axis=0)
normalized = (grades - col_min) / (col_max - col_min)
print("Min-Max normalized (per subject):\n", normalized)
```

```
flattened = grades.ravel()
print("Flattened vector:", flattened)
```
