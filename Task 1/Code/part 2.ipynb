```python
import numpy as np
```

```python
X = np.array([50, 60, 80, 100, 120], dtype=float)
y = np.array([150, 180, 240, 300, 330], dtype=float)
print(X)
print(y)
```

```python
X_col = X.reshape(-1, 1)
print(X_col)
```

```python
m = X_col.shape[0]
X_b = np.hstack([np.ones((m, 1)), X_col])
print(X_b)
```

```python
XtX = X_b.T @ X_b
XtX_inv = np.linalg.inv(XtX)
Xty = X_b.T @ y
theta = XtX_inv @ Xty
print("θ (bias, slope):", theta)
```

```python
x_new = np.array([1.0, 90.0])
pred_90 = float(x_new @ theta)
print("Predicted price for 90 m² (in thousands):", pred_90)
```
